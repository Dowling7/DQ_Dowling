```python
import dwong
```

```python
def gen_wew(file):
    output=dwong.find_coord(file)
    x_eve=output[0]
    y_eve=output[1]
    eng_eve=output[2]
    wew_x = []
    wew_y = []
    for i in range(len(eng_eve)):
        eng_tot=sum(eng_eve[i])
        x_bar=(dwong.np.dot(x_eve[i], eng_eve[i]))/eng_tot
        y_bar=(dwong.np.dot(y_eve[i], eng_eve[i]))/eng_tot
        x_sq_eve = []
        y_sq_eve = []
        for j in range(len(eng_eve[i])):
            x_sq_eve.append(eng_eve[i][j]*(x_eve[i][j]-x_bar)**2)
            y_sq_eve.append(eng_eve[i][j]*(y_eve[i][j]-y_bar)**2)
        try: wew_x.append(dwong.math.sqrt(sum(x_sq_eve)/eng_tot))
        except ZeroDivisionError: wew_x.append(-1)
        try: wew_y.append(dwong.math.sqrt(sum(y_sq_eve)/eng_tot))
        except ZeroDivisionError: wew_y.append(-1)

    return wew_x, wew_y
```
