```python
# Adjusting the histogram code to show the ratio instead of the raw count on the y-axis
import numpy as np
# Placeholder paths for the CSV files
electron_file_path = '/Users/wongdowling/Documents/Github/DQ_Dowling/new_ID/w_Ep/csv_manual_gpz/p5_80_electron_0_10000.csv'  # Replace with actual path
photon_file_path = '/Users/wongdowling/Documents/Github/DQ_Dowling/new_ID/w_Ep/csv_manual_gpz/p5_80_photon_0_10000.csv'      # Replace with actual path

# Load the CSV files
df_electron = pd.read_csv(electron_file_path)
df_photon = pd.read_csv(photon_file_path)

# Splitting the electron dataset into two categories based on 'trkl_x' value
df_electron_minus9999 = df_electron[df_electron['trkl_x'] == -9999]
df_electron_others = df_electron[df_electron['trkl_x'] != -9999]

# List of columns to plot histograms for
columns_to_plot = ['wid_x', 'wid_y', 'wew_x', 'wew_y']

# Creating histograms with y-axis as ratio
for column in columns_to_plot:
    plt.figure(figsize=(10, 6))

    # Total counts for normalization
    total_count_minus9999 = len(df_electron_minus9999)
    total_count_others = len(df_electron_others)
    total_count_photon = len(df_photon)

    # Histogram for electron events with trkl_x = -9999 (normalized)
    plt.hist(df_electron_minus9999[column], bins=30, alpha=0.5, 
             weights=np.ones(total_count_minus9999) / total_count_minus9999,
             log=True,
             label='Electron sample w/o trkl')

    # Histogram for electron events with trkl_x != -9999 (normalized)
    plt.hist(df_electron_others[column], bins=30, alpha=0.5, 
             weights=np.ones(total_count_others) / total_count_others,
             log=True,
             label='Electron sample w/ trkl')

    # Histogram for photon events (normalized)
    plt.hist(df_photon[column], bins=30, alpha=0.5, 
             weights=np.ones(total_count_photon) / total_count_photon,
             log=True,
             label='Photon')

    plt.title(f'Normalized Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Ratio')
    plt.legend()
    plt.savefig(f' Histogram of {column}')
    plt.show()

# Note: The actual execution of this code requires the correct file paths and the files to be accessible.
```
