```python
import numpy as np
import matplotlib.pyplot as plt

# Create theta values
theta = np.linspace(0, 2 * np.pi, 1000)

# Define radii for the concentric circles. For this example, I'll use intervals of 5.
radii = np.arange(5, 41, 10)

# Create a figure
fig = plt.figure(figsize=(8, 8))

# First, set up a standard Cartesian plot (this will show the grid)
ax_cartesian = fig.add_subplot(111)
ax_cartesian.grid(True)
ax_cartesian.set_xlim(-40, 40)
ax_cartesian.set_ylim(-40, 40)
ax_cartesian.set_aspect('equal', 'box')

# Then, overlay a polar plot on top of the Cartesian plot
ax_polar = fig.add_subplot(111, projection='polar', frame_on=False)

# Plot each circle on the polar axes
for r in radii:
    ax_polar.plot(theta, [r] * len(theta))

# Add a star point at the origin (radius 0)
ax_polar.plot(0, 0, 'ro', marker='*', markersize=10)
# Add label "energy seed" at the center
ax_polar.text(5, 0, "energy seed", horizontalalignment='center', verticalalignment='bottom')

# Add other star points
points_theta = [np.radians(45), np.radians(135), np.radians(225), np.radians(315), np.radians(60)]
points_r = [10, 20, 30, 35, 40]

for t, r in zip(points_theta, points_r):
    ax_polar.plot(t, r, 'bo', marker='*', markersize=8)

# Set y ticks to match the radii and remove angle labels for the polar axes
ax_polar.set_yticks(radii)
ax_polar.set_yticklabels([str(r) for r in radii])
ax_polar.set_xticks([])  # remove angle labels
ax_polar.grid(False)  # remove polar grid

plt.title("Seed-Kmean clustering schematic on EMCal")
plt.show()
```
